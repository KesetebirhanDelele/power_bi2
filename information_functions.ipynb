# CONTAINSROW vs IN

> __Note__ that IN{ } is equivalent to CONTAINSROW({ })

> The same is true for NOT [COLOUR] IN{ } is equivalent to NOT CONTAINSROW({ }, [COLOUR])

    EVALUATE
        FILTER (
            ALL ( Product[Color] ),
            ( [Color] )
                IN {
                "Red",
                "Yellow",
                "Blue"
            }
        )
        ORDER BY [Color]

    EVALUATE
        FILTER (
            ALL ( Product[Color] ),
            CONTAINSROW (
                {
                    "Red",
                    "Yellow",
                    "Blue"
                },
                [Color]
            )
        )
        ORDER BY [Color]
