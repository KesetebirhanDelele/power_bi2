# Drill options

## Drill mode: 
click a hierrarchy value to see details at next level for the selected value. Eg IF year 2000 selected, values for year 2000 and quarters for the year will be seen


## Next level: 
all next level values shown Eg quarters


## Expand all down one level: 
values for current and next hierrarchy shown


## Data point table:

> Before: click on a specific value to see records aggregating to that value within the table


> After


Return to before by clicking on Back to Report
