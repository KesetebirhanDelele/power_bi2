# Drill through

Set up drill through feature or visual on a page to be drilled through to using format section 



Make sure keep filters is "On" and the Drill through from section has a column name in this case Region Name:
Note to ctrl+back arrow to return back to starting page


Now go to the page of interest, right click on a specific region name, and click drill through to the page set-up for drill through



## Creating multiple Drill through pages

> After creating the first drill through page, duplicate, rename and change content of new drill thru pages. This is because it doesn't allow you to create more than one drill thru set up more than once. See the drill thru page set up below for the fatalities page below
