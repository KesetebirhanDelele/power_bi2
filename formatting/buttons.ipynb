## Back button

> To set up the back button, first be on the page that you want to go back to after clicking the 'back button'. 

> Then place the back button in the desired page by inserting it and go to action to activate it.

> The default action for 'back' button is 'back'

> To test it, 'ctrl+back' and that will take you back to the previous page
