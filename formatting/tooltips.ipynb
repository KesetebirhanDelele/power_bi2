# Tooltips

> Note that tooltips can give you information you want to give to the user without actually having to use drill through or buttons or actions, making them supirior function.

## Setting up a page to be a tooltip whithin page information setting
> don't click any visual if you want to set the whole page as a tooltip and select page type and filter (show tooltip on option) as indicated

> Seleting 'auto' on tooltip page option will let Power BI desice which tooltip to show



## Linking visual to a tooltip page:
> Click on the visual containing region name then go to properties, turn tooltips on and select tooltip landing page, which is named 'tooltip page' in this case

> You can let the tooltip type to be 'default' if you want PowerBI to show defualt tooltip (which is n't much) 



## Viewing the tooltip by hovering on a visual containing region name
> Note that the region name is automatically selected on the tooltip page since it gets filtered by region name


> Note to hide tooltip page in PowerBI desktop so that it is not seen after the report gets pubished on PowerBI online to avoid redundancy. It should be seen when you hover on region name
