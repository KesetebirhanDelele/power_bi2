# Page and Bookmark navigators

> Note where you can get it: Insert>Buttons>Navigator

> Note the Options to see or hide hidden pages and/or tooltip pages

> Note that current page is highlighted black since we are on it

> You can insert page navigator in all pages to allow navigation from one page to another with ease
