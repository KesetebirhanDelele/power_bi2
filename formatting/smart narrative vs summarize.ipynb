# Smart narrative vs Summarize

> Smart narrative used to summarize all figures in a page

> Summarize used to narrate a summary for a single figure by right clicking on a figure and selecting 'summarize'

> Note that you can click in the summary text and add value for the narrative to include (see option below) by asking a question and saving the response

> You can also edit the response

> You can change the narrative by selecting filter meaning the summary is dynamic and changes based on selection
