# Editing hierarchy



> Note that once hierarchy is created, you should hide elements of a hierarchy so that it doesn't confuse users with duplicate names.

> It can be hidde in model view or data view

# Categoriaing location data in columns



> Note that you can't categorize or reuse a column that is allocated to a hierarchy and hidden away.

> You have to first unhide it, rename it, and edit.

> ALternatively, you can categorize data type by ging into model view, where all hidden or unhidden columns can be seen
